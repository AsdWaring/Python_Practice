```python
import altair as alt

chart = alt.Chart(pokemon).mark_bar().encode(
    x='type',
    y='capture_rate'
)
chart
```

```python
plot_df = pd.DataFrame(pokemon.rename(columns={'capture_rt':'capture_rate'})
.assign(AD_total=pokemon['defense']+ pokemon['attack']))
pokemon_plot = alt.Chart(plot_df, width=500, height=300).mark_circle().encode(
    x='AD_total',
    y='capture_rate'
)
pokemon_plot
```

```python
pokemon_plot2=alt.Chart(plot_df, width=500, height=300).mark_circle(color='green', size=80, opacity=0.3).encode(
    x='AD_total',
    y='capture_rate'
).properties(title='Capture Rate and Attack/Defence Total')
pokemon_plot2
```

If the data isn't running properlly you set the data type in Altair

)
Ordinal= O - descrete or ordered data
Nominal= N - discrete unordered data
Quantitative = Q - continious quantity
Temporal = T - time or date value

Tidy Data!
Each row is a single observation
each variable is a single column
each value is a single cell

Pivoting transforms long dataframes into wider ones.
This is usefull if the dataframe has duplicate rows due to the data not being tidy.

.pivot()

takes the following arguements
index - uesed to make the new dataframes index
columns - used to select which column you want to transform
values - the column name where the values of the column you want to pivot exists

when using pivot we loose the other columns. 

Use .pivot_table() in most cases

also remember to .reset_index()

.rename_axis('') - will remove the index column

.pivot_table() takes averages of duplicate columns

.duplicated() will remove the duplicated data




```python
pokemon_remove_duplicate=pokemon.duplicated(subset='total_bs', keep=False)
pokemon_remove_duplicate
```

```python
pokemon[pokemon_remove_duplicate]
```

bad example since it's a tidy data frame, but if there was data hiding in the 801 rows then this would help us identify it.

The code below pivoted the data so the name of the pokemon is the rows and the columns are now 'type' because we didn't put in a values arguement

```python
pokemon_pivot=pokemon.pivot_table(index='pokemon',columns='type').reset_index()
pokemon_pivot
```

```python
pokemon_pivot.columns
```

```python
pokemon_pivot=pokemon.pivot_table(index='pokemon',columns='type',values='deck_no').reset_index()
pokemon_pivot
```

.melt() Arugements:

id_vars = columns you want to keep
value_vars = columns you want ot melt
var_nam e= name of the new column
value_name = measurements for the new column are housed

This took the data set and melted all the bug and water columns together.



```python
melted_pokemon=(pokemon_pivot.melt(id_vars='pokemon',
                            value_vars=['bug','water'], 
                            var_name='type',
                            value_name=''))
melted_pokemon
```
