Using .agg('count') - or whatever verb you would like to run, only runs the data for that verb and displays that data for all columns. So above, it is counting the number of types, which is the groupby I had created and then listing the counts of those groups.

```python
import pandas as pd
import altair as alt

pokemon = pd.read_csv('data/pokemon.csv')
pokemon_type = pd.DataFrame(pokemon.groupby('type').mean().loc[:, 'attack'])

pokemon_type = pokemon_type.reset_index()

attack_plot = alt.Chart(pokemon_type, width=500,
                        height=300).mark_bar().encode(x=alt.X('type:N', sort='-y',
        title='Pokemon type'), y=alt.Y('attack:Q',
        title='Mean attack score'
        )).properties(title='Mean attack value among Pokemon types')

attack_plot
```

You must use reset_index() to put the column that you grouped by back into the data set.
