arguments for pd.read.

index_col
header
nrows
usecols

index_col is an argument that indicates which column will be acting as the index label.

df = pd.read_csv('cereal.csv', index_col="name")
df.head()

When the table doesn't populate properlly because of the header:

candybars = pd.read_csv('candybars-h.csv', header=2)
candybars

If you only want to load a ceratin set of rows, you can use nrows=number to limit the number of rows that populate


candybars = pd.read_csv('candybars.csv', nrows=7)
candybars

If you only want certain columns use nrows.

you can either use index position or label names. 

candybars = pd.read_csv('candybars.csv', usecols=[0, 1, 10])
candybars


candybars = pd.read_csv('candybars.csv', usecols=['name', 'weight', 'available_canada_america'])
candybars

```python
pokemon_og = pd.read_csv('data/pokemon.csv', nrows=151)
pokemon_og
```

```python
pokemon_og_special_only = pd.read_csv('data/pokemon.csv', nrows=151, usecols=['name', 'sp_attack', 'sp_defense'] )

pokemon_og_special_rename = pokemon_og_special_only.rename(columns= 
{'sp_attack':'special_attack', 'sp_defense':'special_defense'})
pokemon_og_special_rename
```

```python
import altair as alt

chart = alt.Chart(pokemon_og_special_rename).mark_bar().encode(
    x='name',
    y='special_attack'
)
chart
```
